```python
import matplotlib.pyplot as plt   # plotting library
import pandas as pd               # data manipulation and analysis
from psequant import get_pse_data # pse data from PSE edge
```

```python
# Ticker Symbol
stock = "SMPH"

# Date format (yyyy-mm-dd)
beg_date = "2016-01-01"
end_date = "2020-01-01"

df = get_pse_data(stock, beg_date, end_date)
print(df.head())
```

```python
plt.figure(figsize=(10,5))                      # Figure Size in inches
plt.title('Close Price History of '+ stock)                # Title
plt.plot(df['close'])                           # plot
plt.xlabel('Date', fontsize=10)                 # x label
plt.ylabel('Close Price  (PHP)', fontsize = 10) # y label
plt.grid(True)                                  # Show grids
plt.savefig(stock+'_stock.png')                 # Save as .png
plt.show()                                      # Display Figure

```
