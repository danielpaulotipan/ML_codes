```python
"""
To get daily prices of stocks in PSE.

"""

from psequant import get_pse_data # data PSE Edge
import pandas as pd 
import matplotlib.pyplot as plt

# Ticker Symbol
stock = "SMPH"

# Date format (yyyy-mm-dd)
beg_date = "2016-01-01"
end_date = "2020-01-01"

data = get_pse_data(stock, beg_date, end_date)
data.to_csv(stock + '_price.csv')


data.head()
```

```python
"""

Prediction Analysis with Linear Regression (y = mx + b)
             
"""

from sklearn.linear_model import LinearRegression
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle

df = pd.read_csv(stock + '_price.csv')

df = df.reset_index()
prices = df['close'].tolist()
dates = df.index.tolist()
```

```python
dates = np.reshape(dates, (len(dates), 1))
prices = np.reshape(prices, (len(prices), 1))
```

```python
regressor = LinearRegression()
regressor.fit(dates, prices)
```

```python
plt.scatter(dates, prices, color='blue', label= 'Actual Price') #plotting the initial datapoints
plt.plot(dates, regressor.predict(dates), color='black', linewidth=3, label = 'Predicted Price') #plotting the line made by linear regression
plt.title('Linear Regression | Time vs. Price')
plt.legend()
plt.xlabel('Date Integer')
plt.show()
```
